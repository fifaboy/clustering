```python
import numpy as np
import pandas as pd

from os import listdir
from os.path import join

from sklearn.cluster import KMeans

from deterministic import (
    DeterministicKMeans,
    dot,
)

from kmeans_clustering import KMeansClustering
```

```python
%%time
datadir = 'data/'
outdir = 'out/'
NUM_CLUSTERS = [5, 10, 20]
data = []
K = 10
for file in listdir(datadir)[:]:
    path = join(datadir, file)
    df = pd.read_csv(path)
    for num_cluster in NUM_CLUSTERS:
        kmeans = DeterministicKMeans(df, num_cluster)
        kmeans = kmeans.fit()
        labels = kmeans.predict(df)
        for i in range(K):
            row = {}
            row['n_cluster'] = num_cluster
            row['deterministic'] = kmeans.inertia_
            row['det_iter'] = kmeans.num_iters_
            row['k'] = i+1
            kmeans2 = KMeansClustering(init='k-means++', n_clusters=num_cluster)
            kmeans2.fit_predict(df.to_numpy())
            row['kmeans++'] = kmeans2.inertia_
            row['iter'] = kmeans2.n_iter_
            data.append(row)
data = pd.DataFrame(data)
data
```
